```python

#ELE888 - Intelligent Systems, lab 1, Ahmad Fahmy 500913092
import matplotlib
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline 

from sklearn.linear_model import LinearRegression

data = pd.read_csv("https://raw.githubusercontent.com/tofighi/MachineLearning/master/datasets/student_marks.csv")

X = data['Midterm mark'] 
y = data['Final mark']

X = X.values.reshape(-1,1) 
y = y.values.reshape(-1,1)

m = -0.5
b = 0
alph = 0.0001
#alph = 0.1
y2 = m*X+b

plt.scatter(X,y)
#plt.scatter(X, y2)
plt.show()
```

```python
plt.scatter(X,y)
plt.plot(X, y2, color='red', linewidth=1)
plt.show()
```

```python
new_m_vals = [m]

new_b_vals = [b]
E = [(1/len(X) * sum(y - y2))**2]

for i in range(100):
  partial_m = 2/len(X) * sum(-X*(y-(new_m_vals[i]*X+new_b_vals[i])))
  partial_b = 2/len(X) * sum(-(y-(new_m_vals[i]*X+new_b_vals[i])))
  new_m_vals.append(new_m_vals[i] - alph*partial_m)
  new_b_vals.append(new_b_vals[i] - alph*partial_b)
  E.append(1/len(X) * sum((y - (new_m_vals[i+1]*X+new_b_vals[i+1]))**2))

m100 = new_m_vals[100]
b100 = new_b_vals[100]
y3 = m100*X+b100
plt.scatter(X,y)
plt.scatter(X, y3)
plt.show()
```

```python
plt.scatter(range(101), E)
plt.show()
```
